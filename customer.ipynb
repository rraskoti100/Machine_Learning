# K-means Clustering

```python
from sklearn.cluster import KMeans
```

```python
param = []
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init = 'k-means++', random_state = 0)
    kmeans.fit(data)
    param.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), param)
plt.title("ELbow method")
plt.show()
```

```python
model = KMeans(n_clusters = 5, init = "k-means++", random_state = 0)
model_kmeans = model.fit_predict(data)
```

```python
plt.scatter(data[model_kmeans == 0, 0], data[model_kmeans == 0,1], s = 100, c= "red", label ="Cluster 1")
plt.scatter(data[model_kmeans == 1, 0], data[model_kmeans == 1,1], s = 100, c= "green", label ="Cluster 2")
plt.scatter(data[model_kmeans == 2, 0], data[model_kmeans == 2,1], s = 100, c= "blue", label ="Cluster 3")
plt.scatter(data[model_kmeans == 3, 0], data[model_kmeans == 3,1], s = 100, c= "black", label ="Cluster 4")
plt.scatter(data[model_kmeans == 4, 0], data[model_kmeans == 4,1], s = 100, c= "cyan", label ="Cluster 5")
plt.title("cluster of customers")
plt.xlabel("Annual income (k$)")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
